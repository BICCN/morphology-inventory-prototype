```python
import os
import pandas as pd
```

```python
#
# setup input directory
# 
input_directory = 'C:/Users/lydia/OneDrive/Work/DigitalAssets/BICCNMorphologyInventory/u19zeng'
```

```python
sample_file = 'sample_inventory_with_directory_info.csv'
downsample_file = 'downsampled_images_inventory.csv'
reconstruction_file = 'reconstruction_inventory.csv'
```

```python
#
# setup output directory
# 
output_directory = 'C:/Users/lydia/OneDrive/Work/DigitalAssets/BICCNMorphologyInventory/u19zeng'
```

```python
#
# open input files as dataframes
#
samples = pd.read_csv(os.path.join(input_directory,sample_file))
downsamples = pd.read_csv(os.path.join(input_directory,downsample_file))
recons = pd.read_csv(os.path.join(input_directory,reconstruction_file))
```

```python
samples.set_index('HUST Sample ID', inplace=True)
downsamples.set_index('Filename', inplace=True)
recons.set_index('Reconstruction ID', inplace=True)
```

```python
count = 0

for s, rows in samples.iterrows() :
    
    print( str(count) + ": " + str(s) )
    
    df = downsamples[downsamples['HUST Sample ID'] == str(s)]
    
    if len(df) > 0 :
        
        val = df.index.values
        samples.at[s,'Downsample Files'] = ';'.join(val)
        
        df = df[df['Image Type'] == '.tif']
        df = df[df['XY Downsample'] == 32 ]
        df = df[df['Z Downsample'] == 8 ]
        
        cf = df[df['Channel'] == 'red']
        if len(cf) > 0 :
            samples.at[s, 'Red Downsample File'] = cf.index.values[0]
            
        cf = df[df['Channel'] == 'green']
        if len(cf) > 0 :
            samples.at[s, 'Green Downsample File'] = cf.index.values[0]

    
    df = recons[recons['HUST Sample ID'] == s]
    
    if len(df) > 0 :
        
        df = df[df['Reconstruction Type'].isna()]
        samples.at[s, 'Number of Reconstructions'] = len(df)
        
    count = count + 1
```
