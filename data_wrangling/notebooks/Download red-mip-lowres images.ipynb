```python
import os
import pandas as pd
import numpy as np
import ftplib
```

```python
inventory_directory = r'C:\Users\LydiaN\OneDrive - Personal\OneDrive\Work\DigitalAssets\BICCNMorphologyInventory\u19zeng'
inventory_file = os.path.join( inventory_directory, 'file_inventory_part_2.csv')
inventory = pd.read_csv( inventory_file )
inventory.set_index('sample_id', inplace=True)

p1 = inventory['channel'] == 'red'
p2 = inventory['image_type'] == 'mip'
p3 = inventory['resolution'] == 'lowres'
pred = np.logical_and( p1, p2 )
pred = np.logical_and( pred, p3 )
filtered = inventory[pred]
print(len(filtered))
```

```python
def get_dir_info(ftp) :
    
    data = []
    ftp.dir(data.append)
    data = [x.split() for x in data if x != '']
    data = pd.DataFrame(data)
    
    return data

def getFile(ftp, base_path, filename):
    try:
        ftp.retrbinary("RETR " + filename ,open(os.path.join(base_path,filename), 'wb').write)
    except:
        print("Error")
```

```python
skip_list = [18470,18469,18468,18049,18047,18054,192337,192338,192339,196478,182709,192343]
request_list = []
```

```python
for iindex, irow in filtered.iterrows() :
    
    if irow['hust_imaging_id'] in skip_list :
        continue
        
    if len(request_list) > 0 and irow['hust_imaging_id'] not in request_list :
        continue
    
    ftp = ftplib.FTP(host="download.brainimagelibrary.org")
    ftp.login()
    
    print(iindex)
    home_dir = ftp.pwd()
    remote_base = "biccn/zeng/luo/fMOST"
    
    # get listing of remote directory
    rpath = remote_base + "/" + iindex + "/red/mip/lowres"
    ftp.cwd( rpath )
    
    # get local path
    lpath = os.path.join( inventory_directory, iindex, 'red', 'mip', 'lowres')
    os.makedirs(lpath, exist_ok = True)
    
    # copy tif files in directory
    data = get_dir_info(ftp)
    pred = ['.tif' in x for x in data[8]]
    tf = data[pred]
    print( "- copying %d tif files" % len(tf))
    for findex, frow in tf.iterrows() :
        getFile( ftp, lpath, frow[8])
        
    ftp.quit()
    
```
