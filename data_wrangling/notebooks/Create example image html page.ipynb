```python
import os
import pandas as pd

import dominate
from dominate.tags import *
```

```python
input_directory = r'\\allen\programs\celltypes\workgroups\mousecelltypes\_UPENN_fMOST\shared\morphology_inventory\hust_imaging_example_thumbnails'
#input_directory = r'C:\Users\lydia\OneDrive\Work\DigitalAssets\BICCNMorphologyInventory\u19zeng\hust_imaging_example_thumbnails'
inventory_path = os.path.join(input_directory,'manifests','examples_with_orientation.csv')
inventory = pd.read_csv(inventory_path)
inventory = inventory.sort_values(by = 'hust_imaging_id')
```

```python
title = "fMOST example images"
doc = dominate.document(title=title)
doc += h1(title)

doc.head += link(href="reconstruction_table.css",rel="stylesheet", type="text/css", media="all")

div = div(id='hust_imaging')
with div :
    
    input_(cls="search",placeholder="Search")
    
    with table().add(tbody(cls="list")):

        for rrindex, rrow in inventory.iterrows() :
            
            l = tr()
            
            mlist = ['hust_imaging_id', 'labtrack_mouse_id','full_resolution_image_path', 'x_size','y_size','orientation']
            for m in mlist :
                l += td(rrow[m],cls=m)
                
            tpath = os.path.join(input_directory,rrow['downsampled_image_path'])
            tpath = tpath.replace("\\", "/")
            tpath = "file://" + tpath
            
            l += td(img(src=tpath,height="300"))

                    
doc.body += div
doc.body += script(src="https://cdnjs.cloudflare.com/ajax/libs/list.js/1.5.0/list.min.js")

jcode = \
'''
var options = {
  valueNames: [ 'hust_imaging_id','orientation' ]
};

var userList = new List('hust_imaging', options);
'''

doc.body += script(jcode)
```

```python
output_directory = r'\\allen\programs\celltypes\workgroups\mousecelltypes\_UPENN_fMOST\shared\morphology_inventory'
#output_directory = r'C:\Users\lydia\OneDrive\Work\morphology_tracking\examples'
output_file = os.path.join( output_directory,'hust_imaging_inventory.html')
with open(output_file, 'w') as f :
    f.write(doc.render())
```
