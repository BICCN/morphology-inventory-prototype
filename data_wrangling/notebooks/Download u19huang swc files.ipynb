```python
import os
import pandas as pd
import ftplib
```

```python
def getFile(ftp, base_path, filename):
    try:
        ftp.retrbinary("RETR " + filename ,open(os.path.join(base_path,filename), 'wb').write)
    except:
        print("Error")
        
def get_dir_info(ftp) :
    
    data = []
    ftp.dir(data.append)
    data = [x.split() for x in data if x != '']
    data = pd.DataFrame(data)
    
    return data

def check_for_directory(data,dirname) :
    
    has_dir = False
    for dindex, drow in data.iterrows() :
        if drow[8] == dirname :
            has_dir = True
    return has_dir
```

```python
base_directory = r'C:\Users\LydiaN\OneDrive - Personal\OneDrive\Work\DigitalAssets\BICCNMorphologyInventory\u19huang\cell_datasets'
collection = '3ac1bdc022d0da78'

output_directory = os.path.join( base_directory, collection )

remote_base = collection[0:2] + "/" + collection[2:4] + "/" + collection
```

```python
ftp = ftplib.FTP(host="download.brainimagelibrary.org")
ftp.login()

home_dir = ftp.pwd()
ftp.cwd( remote_base )
directories = get_dir_info( ftp )

pred = [x not in ['.','..'] for x in directories[8]]
fdirs = directories[pred]

for findex, frow in fdirs.iterrows() :
    
    sdir = frow[8]
    print(sdir)
    
    # make local directory
    ldir = os.path.join( output_directory, sdir )
    os.makedirs( ldir, exist_ok = True )
    
    # cd to remote directory
    ftp.cwd( home_dir + "/" + remote_base + "/" + sdir )
    swcs = get_dir_info( ftp )
    pred = ['.swc' in x for x in swcs[8]]
    fswcs = swcs[pred]
    
    for sindex, srow in fswcs.iterrows() :
        getFile( ftp, ldir, srow[8] )
    
    print("-- done ")

ftp.quit()
```
