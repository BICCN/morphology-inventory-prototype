```python
import pandas as pd
import dominate
from dominate.tags import *
```

```python
input_file = 'Supplemental_Table_2_41586_2021_3941_MOESM4_ESM.xlsx'
df = pd.read_excel(input_file,header=1)
```

```python
df.head()
```

```python
output_file = 'index.html'
title = "u19zeng registered reconstructions"
```

```python
doc = dominate.document(title=title)
doc += h1(title)

doc.head += link(href="reconstruction_table.css",rel="stylesheet", type="text/css", media="all")

div = div(id='reconstructions')
with div :
    
    input_(cls="search",placeholder="Search")
    
    with table().add(tbody(cls="list")):

        for rrindex, rrow in df.iterrows() :
            
            l = tr()
            
            mlist = ['formatted name','Manually_corrected_soma_region']
            for m in mlist :
                l += td(rrow[m],cls=m)
                                   
            tpath = "%s_ccf_combined_view.png" % (rrow['formatted name'])
            tpath = "images/" + tpath
            
            l += td(img(src=tpath,height="300"))
                        
            tpath = "%s_ccf_combined_atlas_colors.png" % (rrow['formatted name'])
            tpath = "images/" + tpath
            
            l += td(img(src=tpath,height="300"))
 
                    
doc.body += div
doc.body += script(src="https://cdnjs.cloudflare.com/ajax/libs/list.js/1.5.0/list.min.js")
```

```python
jcode = \
'''
var options = {
  valueNames: [ 'formatted name', 'Manually_corrected_soma_region' ]
};

var userList = new List('reconstructions', options);
'''

doc.body += script(jcode)
```

```python
with open(output_file, 'w') as f :
    f.write(doc.render())
```
