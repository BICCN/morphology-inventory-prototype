```python
import pandas as pd
import dominate
from dominate.tags import *
```

```python
input_file = '202205111930_upload_mouselight_reconstructions.csv'
df = pd.read_csv(input_file)
```

```python
df.head()
```

```python
output_file = 'index.html'
title = "mouselight reconstructions"
```

```python
doc = dominate.document(title=title)
doc += h1(title)

doc.head += link(href="reconstruction_table.css",rel="stylesheet", type="text/css", media="all")

div = div(id='reconstructions')
with div :
    
    input_(cls="search",placeholder="Search")
    
    with table().add(tbody(cls="list")):

        for rrindex, rrow in df.iterrows() :
            
            l = tr()
            
            mlist = ['cell_name','structure_label']
            for m in mlist :
                l += td(rrow[m],cls=m)
                                   
            tpath = "1_%s_ccf_combined_view.png" % (rrow['cell_name'])
            tpath = "images/" + tpath
            
            l += td(img(src=tpath,height="300"))
                        
            tpath = "1_%s_ccf_combined_atlas_colors.png" % (rrow['cell_name'])
            tpath = "images/" + tpath
            
            l += td(img(src=tpath,height="300"))
 
                    
doc.body += div
doc.body += script(src="https://cdnjs.cloudflare.com/ajax/libs/list.js/1.5.0/list.min.js")
```

```python
jcode = \
'''
var options = {
  valueNames: [ 'cell_name', 'structure_label' ]
};

var userList = new List('reconstructions', options);
'''

doc.body += script(jcode)
```

```python
with open(output_file, 'w') as f :
    f.write(doc.render())
```
